```python
import pandas as pd
import numpy as np
import scipy.stats as stats
```

```python
impressions_path= "impressions.json"
impressions_raw= pd.read_json(impressions_path)

impressions_raw.drop_duplicates(inplace=True)
nan_value = float("NaN")
impressions_raw.replace("", nan_value, inplace=True)
impressions_raw.dropna(subset = ["country_code"], inplace=True)
impressions = impressions_raw[impressions_raw["country_code" and "id"].str.contains("None" or "NaN")==False]
impressions_raw = impressions_raw.astype({"app_id": int, "advertiser_id": int, "country_code": str, "id": str})
impressions
```

```python
clicks_path= "clicks.json"
clicks= pd.read_json(clicks_path)
clicks = clicks.rename(columns={"impression_id": "id"})
clicks.drop_duplicates(inplace=True)
clicks = clicks.astype({"id": str, "revenue": float})
clicks
```

```python
merged = pd.merge(impressions, clicks, on="id")
merged
```

```python
impression_count = impressions.groupby(['app_id', 'country_code']).size().reset_index(name='impression_count')
impression_count
```

```python
click_count = clicks.groupby(['id']).size().reset_index(name='click_count')
click_count
```

```python
revenue = merged.groupby(['country_code','app_id']).sum().reset_index()
revenue = revenue[['country_code', 'app_id', 'revenue']]
revenue
```
