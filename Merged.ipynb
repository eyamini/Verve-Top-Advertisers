```python
import pandas as pd
import numpy as np
import scipy.stats as stats
```

```python
impressions_path= "impressions.json"
impressions= pd.read_json(impressions_path)
impressions.head()
```

```python
clicks_path= "clicks.json"
clicks= pd.read_json(clicks_path)
clicks.head()
```

```python
clicks = clicks.rename(columns={"impression_id": "id"})
clicks.head()
```

```python
merged = pd.merge(impressions, clicks, on="id")
merged
```

```python
merged.drop_duplicates(inplace=True)
```

```python
merged
```

```python
nan_value = float("NaN")
merged.replace("", nan_value, inplace=True)
merged.dropna(subset = ["country_code"], inplace=True)
```

```python
cleaned = merged[merged["country_code"].str.contains("None" or "NaN")==False]
```
