```python
import pandas as pd
import numpy as np
import scipy.stats as stats
```

```python
impressions_path= "impressions.json"
impressions= pd.read_json(impressions_path)

impressions.drop_duplicates(inplace=True)
nan_value = float("NaN")
impressions.replace("", nan_value, inplace=True)
impressions.dropna(subset = ["country_code"], inplace=True)
impressions = impressions[impressions["country_code" and "id"].str.contains("None" or "NaN")==False]
impressions = impressions.astype({"app_id": int, "advertiser_id": int, "country_code": str, "id": str})
impressions
```

```python
clicks_path= "clicks.json"
clicks= pd.read_json(clicks_path)

clicks = clicks.rename(columns={"impression_id": "id"})
clicks = clicks.astype({"id": str, "revenue": float})
clicks
```

```python
merged = pd.merge(impressions, clicks, on="id")
merged
```

```python
impression_count = merged.groupby(['country_code','app_id','advertiser_id'])['id'].size().reset_index(name='impression_count')
impression_count = impression_count.astype({'impression_count': int})
impression_count
```

```python
click_count = merged.groupby(['country_code','app_id','advertiser_id'])['revenue'].size().reset_index(name='click_count')
click_count
```

```python
revenue = merged.groupby(['country_code','app_id','advertiser_id'])['revenue'].sum().reset_index()
revenue
```

```python
metric=pd.merge(impression_count, revenue)
metric
```

```python
metrics=pd.merge(metric, click_count)
```

```python
metrics= metrics.rename(columns={"impression_count": "impressions", "click_count": "clicks"})
metrics= metrics[['app_id', 'country_code', 'advertiser_id', 'impressions', 'clicks', 'revenue']]
```

```python
metrics["rate_per_click"] = metrics["revenue"]/metrics["impressions"]
metrics
```

```python
top_5=metrics.nlargest(36,['rate_per_click'])
top_5=top_5[['app_id', 'country_code', 'advertiser_id']]
top_5['recommended_advertiser_ids']=top_5.advertiser_id.tolist()
top_5.sort_values(['app_id', 'country_code'], ascending=[True, False], inplace=True)
```

```python
top_5=top_5[['app_id', 'country_code', 'recommended_advertiser_ids']]
top_5
```

```python
top_5.to_json('recommended_advertiser_ids.json', orient='records')
```
